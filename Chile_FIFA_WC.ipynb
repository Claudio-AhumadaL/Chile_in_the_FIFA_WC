```python
import pandas as pd
```

```python
# Load Teams database

teams = pd.read_csv('./data-csv/teams.csv')
teams.head()
```

```python
# Restrict to only have values for Chile. This will be updated later to hold the values of teams that played against Chile

teams_chile = teams[teams['team_name'] == 'Chile']
teams_chile
```

```python
# Load bookings database and restrict to only have values for Chile

bookings = pd.read_csv('./data-csv/bookings.csv')
bookings_chile = bookings[bookings['team_name'] == 'Chile']
bookings_chile
```

```python
# Load goals database and obtain two different dataframes, one for goals Scored by Chile and another for goals scored to Chile

goals = pd.read_csv('./data-csv/goals.csv')
goals_by_chile = goals[goals['team_name'] == 'Chile']

mask = goals['match_name'].str.contains('Chile') & ~goals['team_name'].str.contains('Chile')
goals_to_chile = goals[mask]
```

```python
# Load group_standings database and obtain a dataframe for Chile's results

group_standings = pd.read_csv('./data-csv/group_standings.csv')

group_standings_chile = group_standings[group_standings['team_name'] == 'Chile']
group_standings_chile.head()
```

```python
# Load manager_appearances database and obtain a dataframe for Chile's managers

manager_appearances = pd.read_csv('./data-csv/manager_appearances.csv')

manager_appearances_chile = manager_appearances[manager_appearances['team_name'] == 'Chile']
manager_appearances_chile.head()
```

```python
# Load matches database and obtain a dataframe for Chile's matches

matches = pd.read_csv('./data-csv/matches.csv')

matches_mask = matches['match_name'].str.contains('Chile')
matches_chile = matches[matches_mask]

matches_chile.head()
```

```python
# Load player_appearances database and obtain a dataframe for Chile's players

player_appearances = pd.read_csv('./data-csv/player_appearances.csv')

player_appearances_chile = player_appearances[player_appearances['team_name'] == 'Chile']

player_appearances_chile.head()
```
